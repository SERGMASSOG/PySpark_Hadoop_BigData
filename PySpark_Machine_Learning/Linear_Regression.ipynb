````python
import pandas as pd
from sklearn.linear_model import LinearRegression
````

````python
# uusar pandas para cargar el conjunto de datos en un dataframe.
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-BD0231EN-SkillsNetwork/datasets/mpg.csv"
df = pd.read_csv(URL)
````

````python
df.sample(5)
````

````python
df.shape
````

````python
df.plot.scatter(x = "Horsepower", y = "MPG")
````

## Análisis Exploratorio: Relación entre Horsepower y MPG
El gráfico de dispersión muestra la relación entre la potencia del motor (Horsepower) y la eficiencia de combustible (Miles Per Gallon, MPG). Cada punto azul representa un vehículo en el conjunto de datos.

## Observaciones principales
- Existe una correlación negativa clara: a medida que aumenta la potencia del motor, la eficiencia de combustible tiende a disminuir.
- Los vehículos con alto horsepower suelen presentar valores de MPG bajos, lo que refleja un mayor consumo de combustible.
- En el rango de baja potencia, la variabilidad de MPG es mayor, indicando que otros factores (peso del vehículo, aerodinámica, tipo de transmisión) también influyen en la eficiencia


## Separacion de variables para entrenamiento y creacion de un modelo

````python
target = df["MPG"]
````

````python
features = df[["Horsepower","Weight"]]
````

## Linear Regression Model

````python
lr = LinearRegression()
````

````python
lr.fit(features,target)
````

## Evaluacion del modelo y prediccion

````python
lr.score(features,target)
````

````python
lr.predict([[100,2000]])
````

````python
URL2 = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-BD0231EN-SkillsNetwork/datasets/diamonds.csv"
````

````python
df2 = pd.read_csv(URL2)
````

````python
df2.head()
````

````python
target = df2["price"]
features = df2[["carat","depth"]]
````

````python
lr2 = LinearRegression()
lr2.fit(features,target)
````

````python
#your code goes here
lr2.score(features,target)
````

````python
lr2.predict([[0.3, 60]])
````
